```python
import pandas as pd
import os

datasets_folder = '/content/datasets/'

dfs = []

for filename in os.listdir(datasets_folder):
    if filename.endswith('.csv'):
        filepath = os.path.join(datasets_folder, filename)
        df = pd.read_csv(filepath)
        dfs.append(df)

if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
```

```python
new_column_names = {
    'ФИО': 'full_name',
    'Желаемая должность': 'desired_position',
    'Зарплата': 'salary',
    'Личная информация': 'personal_info',
    'Местоположение': 'location',
    'Занятость и график': 'employment_and_schedule',
    'Общий опыт': 'total_experience',
    'Последнее место работы': 'last_job',
    'Обязанности': 'responsibilities',
    'Ключевые навыки': 'key_skills',
    'Обо мне': 'about_me',
    'Образование': 'education',
    'Языки': 'languages',
    'Ссылка': 'link'
}

combined_df.rename(columns=new_column_names, inplace=True)
```

```python
combined_df.head()
```

```python
combined_df[['last_position', 'last_job_company']] = combined_df['last_job'].str.split(' в ', n=1, expand=True)
```

```python
combined_df.head()
```

```python
combined_df.drop('last_job', axis=1, inplace=True)

columns = combined_df.columns.tolist()

total_experience_index = columns.index('total_experience')
columns.remove('last_position')
columns.remove('last_job_company')
columns.insert(total_experience_index + 1, 'last_position')
columns.insert(total_experience_index + 2, 'last_job_company')

combined_df = combined_df[columns]
```

```python
combined_df.info()
```

```python
combined_df.drop_duplicates(subset=['link'], inplace=True)
```

```python
combined_df.to_csv('combined_df.csv', index=False)
```

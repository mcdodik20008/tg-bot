```python
import pandas as pd
import os

datasets_folder = 'tg-bot/datasets/resume'

dfs = []

for filename in os.listdir(datasets_folder):
    if filename.endswith('.csv'):
        filepath = os.path.join(datasets_folder, filename)
        df = pd.read_csv(filepath)
        dfs.append(df)

if dfs:
    combined_df = pd.concat(dfs, ignore_index=True)
```

```python
new_column_names = {
    'ФИО': 'full_name',
    'Желаемая должность': 'desired_position',
    'Зарплата': 'salary',
    'Личная информация': 'personal_info',
    'Местоположение': 'location',
    'Занятость и график': 'employment_and_schedule',
    'Общий опыт': 'total_experience',
    'Последнее место работы': 'last_job',
    'Обязанности': 'responsibilities',
    'Ключевые навыки': 'key_skills',
    'Обо мне': 'about_me',
    'Образование': 'education',
    'Языки': 'languages',
    'Ссылка': 'link'
}

combined_df.rename(columns=new_column_names, inplace=True)
```

```python
combined_df.drop_duplicates(subset=['link'], inplace=True)
combined_df.reset_index(drop=True)
```
